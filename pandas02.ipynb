```python
temblores_espana_mayores7 = earthquakes2.loc[(earthquakes2['country'] == 'Spain') | (earthquakes2['magnitude'] >= 9.0) ]
temblores_espana_mayores7
```

```python
temblores_en_america_norte = earthquakes2.loc[earthquakes2['country'].isin(['Mexico', 'Canada', 'United States of America']) ]
temblores_en_america_norte.shape
```

```python
temblores_en_america_norte[temblores_en_america_norte['country'] == 'Mexico'].shape
```

```python
temblores_con_pais = earthquakes2.loc[earthquakes2['country'].notnull()]
temblores_con_pais.shape
```

```python
temblores_sin_pais = earthquakes2.loc[earthquakes2['country'].isnull(), ['location', 'country']]
temblores_sin_pais
```
