```python
import pandas as pd

# Cargar datos (ajusta la ruta si es necesario)
df = pd.read_csv("disasters.csv", parse_dates=["date"])

# Ver las primeras filas
df.head()
```

```python
cols_interes = [
    "severity_index",
    "casualties",
    "economic_loss_usd",
    "response_time_hours",
    "aid_amount_usd",
    "response_efficiency_score",
    "recovery_days",
]

df[cols_interes].describe()
```

```python
paises=df["country"].value_counts()
print(paises)
df["disaster_type"].value_counts(normalize=True) * 100
```

```python
casualties_by_type = df.groupby("disaster_type")["casualties"].agg(["count", "sum", "mean"])
casualties_by_type.sort_values(by="sum", ascending=False)
```
